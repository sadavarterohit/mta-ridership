```python
df = dd.read_csv("drive/MyDrive/Colab Notebooks/MTA_Subway_Hourly_Ridership_Data.csv", dtype={'station_complex_id': 'object'}, parse_dates=['transit_timestamp'])
```

```python
df.head()
```

```python
df_new = df.map_partitions(cudf.from_pandas)
df_new = df[df['transit_mode'] == 'subway']
df_new['year'] = df['transit_timestamp'].apply(lambda x: x.year, meta=('transit_timestamp', 'int64'))
df_new['month'] = df['transit_timestamp'].apply(lambda x: x.month, meta=('transit_timestamp', 'int64'))
df_new['day_type'] = df['transit_timestamp'].apply(lambda x: 2 if x.weekday() in [5, 6] else 1, meta=('transit_timestamp', 'int64'))
df_new = df_new[['year', 'month', 'day_type', 'station_complex', 'ridership']]
df_grouped = df_new.groupby(by=['station_complex', 'year', 'month', 'day_type']).sum()
```

```python
df_grouped.head()
```

```python
df_grouped.to_csv('MTA_Subway_Hourly_Ridership_Data_Modified.csv')
```

```python
import pandas as pd
df_list = []
for i in range(87):
  df_i = dd.read_csv("MTA_Subway_Hourly_Ridership_Data_Modified.csv/{:02d}.part".format(i))
  df_list.append(df_i.compute())
merged_df = pd.concat(df_list)
merged_df.to_csv("MTA_Subway_Hourly_Ridership_Data_Modified_v2.csv")
```

```python
import pandas as pd

file_path = '../data/MTA_Subway_Hourly_Ridership_Data_Modified_v2.csv'

data = pd.read_csv(file_path)
data.head()
```

```python
filtered_data = data[data['year'] > 2021]

filtered_data['day_type'] = filtered_data['day_type'].replace({1: 'weekday', 2: 'weekend'})

filtered_data['lines'] = filtered_data['station_complex'].str.extract(r'\(([^)]+)\)$')

aggregated_data = filtered_data.groupby(['station_complex', 'month', 'day_type'], as_index=False).agg(
    ridership=('ridership', 'mean'),
    lines=('lines', 'first')  
)
```

```python
aggregated_data.to_csv('../data/aggregated_data.csv', index=False)
```
